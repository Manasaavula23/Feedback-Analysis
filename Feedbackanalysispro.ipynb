Pre-processing the data

````
def preprocess(df):
  data = df     #copying the data frame to a temporary variable
  #converting the comments to string type because they are stored as objects in pandas dataframe
  data['comments'] = data['comments'].astype(str)

  #remove all the special characters from the feedbacks
  data['comments'] = data['comments'].apply(lambda x: re.sub('[^A-Za-z]', ' ', x))

  #convert all the feedbacks to lower case and lemmatize the word to find out the meaningful root word from the feedbacks
  data['comments'] = data['comments'].apply(lambda x: " ".join([lmtzr.lemmatize(spell(reduce_lengthening(word.lower().replace(r'^pls', 'please').replace(r'^plz', 'please').replace(r'^u$','you').replace(r'^ur$','your')))) if word.lower() not in mess_items else word.lower() for word in x.split()]))

  #remove the stop words
  data['comments'] = data['comments'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

  #aplly sentiment analysis to all the comments using sentiment function in TextBlob library
  data['sentiment'] = data['comments'].apply(lambda x: TextBlob(x).sentiment[0])
  print('preprocessing complete')
  return data
````

Following code shows how to convert a corpus into a document-term matrix.

**Alpha and Beta Hyperparameters** – alpha represents document-topic density and Beta represents topic-word density.

**Number of Topic Terms** – Number of terms composed in a single topic.

**Number of Iterations / passes** – Maximum number of iterations allowed to LDA algorithm for convergence.

applying the kmeans and visualizing the clusters for each mess separately.

````
%matplotlib inline
neg_data
r = lambda: random.randint(0,255)
KMCampus = {}
# neg_data.keys()
for campusName in neg_data:
  KMCampus[campusName] = {}
  for messName in neg_data[campusName].keys():
    true_k = math.ceil((len(neg_data[campusName][messName])/2)**0.5)
    print(true_k)
    vectorizer = TfidfVectorizer(stop_words='english')
    model = KMeans(n_clusters=true_k, init='k-means++', max_iter=1000, n_init=1)
    X = vectorizer.fit_transform(neg_data[campusName][messName]['comments'])
    model.fit(X)
    order_centroids = model.cluster_centers_.argsort()[:, ::-1]
    terms = vectorizer.get_feature_names()
    clusters = model.labels_.tolist()
    dist = 1 - cosine_similarity(X)
    KMCampus[campusName][messName] = {}
    KMCampus[campusName][messName]['topTerms'] = [[terms[ind] for ind in order_centroids[i, :3]] for i in range(true_k)]
    print(KMCampus[campusName][messName]['topTerms'])
    KMCampus[campusName][messName]['counts'] = list(pd.Series(clusters).value_counts())
    print(KMCampus[campusName][messName]['counts'])
    comments = {'comments':neg_data[campusName][messName]['comments'], 'cluster':clusters}
    cdf = pd.DataFrame(comments, columns=['cluster', 'comments'])
    cdf.index = [clusters]
    KMCampus[campusName][messName]['counts'] = cdf['cluster'].value_counts()
    MDS()
    mds = MDS(n_components=2, dissimilarity="precomputed", random_state=1)
    pos = mds.fit_transform(dist)  # shape (n_components, n_samples)
    xs, ys = pos[:, 0], pos[:, 1]
    cluster_colors = {i: '#%02X%02X%02X' % (r(),r(),r()) for i in range(true_k)}
    cluster_names = {i: str(i) for i in range(true_k)}

    ndf = pd.DataFrame(dict(x=xs, y=ys, label=clusters, title=neg_data[campusName][messName]['comments']))

    groups = ndf.groupby('label')


    fig, ax = plt.subplots(figsize=(17, 9)) # set size
    ax.margins(0.05) # Optional, just adds 5% padding to the autoscaling

    for name, group in groups:
        ax.plot(group.x, group.y, marker='o', linestyle='', ms=12,
                label=cluster_names[name], color=cluster_colors[name],
                mec='none')
        ax.set_aspect('auto')
        ax.tick_params(\
            axis= 'x',          # changes apply to the x-axis
            which='both',      # both major and minor ticks are affected
            bottom='off',      # ticks along the bottom edge are off
            top='off',         # ticks along the top edge are off
            labelbottom='off')
        ax.tick_params(\
            axis= 'y',         # changes apply to the y-axis
            which='both',      # both major and minor ticks are affected
            left='off',      # ticks along the bottom edge are off
            top='off',         # ticks along the top edge are off
            labelleft='off')

    ax.legend(numpoints=1)  #show legend with only 1 point



    plt.show()
````
